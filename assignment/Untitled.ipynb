```python
#import libariry

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
#import dstaset
a= pd.read_csv("ford.csv")
print(a)
```

```python
#check length of dataset
a.head()
```

```python
a.tail()
```

```python
#check datatypes of datasets
a.dtypes
```

```python
#convert the value of tramission coloum data to (0,1,2)
a['transmission'].replace(['Automatic','Manual','Semi-Auto'],[0,1,2],inplace=True)
a.head()
```

```python
#convert the value of tramission coloum data to (0,1,2)
a['fuelType'].replace(['Petrol','Diesel','Hybrid','Electric'],[0,1,2,3],inplace=True)
a
```

```python
#change the datatypes
c_dtypes = {'transmission': int,'fuelType': int }
a = a.astype(c_dtypes)
a.dtypes
```

```python
#find the null values
pd.isnull(a).sum().sum()
```

```python
#find duplicate rows
duplicate = a[a.duplicated()]
duplicate 
```

```python
# delete the null and duplicates values
b=a.dropna().drop_duplicates()
b
```

```python
b.dtypes
```

```python
X = b.drop(['price','model'], axis=1)
y = b['price']

print('Shape of X = ', X.shape)
print('Shape of y = ', y.shape)
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=51)

print('Shape of X_train = ', X_train.shape)
print('Shape of y_train = ', y_train.shape)
print('Shape of X_test = ', X_test.shape)
print('Shape of y_test = ', y_test.shape)
```

```python

sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python

lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
X_test[0, :]
```

```python
lr.predict([X_test[0, :]])
```

```python
lr.predict(X_test)
```
